# **Анализ социальных, экономических и экологических показателей по странам за 2023 год**

Описание признаков:

**Country** - Название страны.

**Density (P/Km2)** - Плотность населения, измеряемая в чел. на квадратный километр.

**Abbreviation** - Сокращение или код, обозначающий страну.

**Agricultural Land (%)** - Процентная доля площади земель, используемых в сельскохозяйственных целях.

**Land Area (Km2)** - Общая площадь страны в квадратных километрах.

**Armed Forces Size** - Численность вооруженных сил страны.

**Birth Rate** - Число рождений на 1000 человек населения в год.

**Calling Code** - Международный телефонный код страны.

**Capital/Major City** - Название столицы или крупного города.

**CO2 Emissions** - Выбросы углекислого газа в тоннах.

**CPI** - Индекс потребительских цен, показатель инфляции и покупательной способности.

**CPI Change (%)** - Процентное изменение индекса потребительских цен по сравнению с предыдущим годом.

**Currency_Code** - Код валюты, используемый в стране.

**Fertility Rate** - среднее число детей, рожденных женщиной в течение ее жизни.

**Forested Area (%)** - Процентная доля территории, покрытой лесами.

**Gasoline_Price** - Цена бензина за литр в местной валюте.

**GDP** - Валовой внутренний продукт, общая стоимость товаров и услуг, произведенных в стране.

**Gross Primary Education Enrollment (%)** - Общий коэффициент охвата начальным образованием.

**Gross Tertiary Education Enrollment (%)** - Общий коэффициент охвата высшим образованием.

**Infant Mortality** - Число смертей на 1000 новорожденных в возрасте до одного года.

**Largest City** -  Название крупнейшего города страны.

**Life Expectancy** - Среднее количество лет, которое, как ожидается, проживет новорожденный.

**Maternal Mortality Ratio** - Число материнских смертей на 100 000 новорожденных.

**Minimum Wage** - Минимальный уровень заработной платы в долларах.

**Official Language** - Официальные языки, на которых говорят в стране.

**Out of Pocket Health Expenditure (%)** - Процентная доля от общих расходов на здравоохранение, оплачиваемых частными лицами из собственных средств.

**Physicians per Thousand** - Количество врачей на тысячу человек.

**Population** - Общая численность населения страны.

**Population: Labor Force Participation (%)** - Процентная доля населения, являющегося частью рабочей силы.

**Tax Revenue (%)** - Налоговые поступления в процентах от ВВП.

**Total Tax Rate** - Общая налоговая нагрузка в процентах от коммерческой прибыли.

**Unemployment Rate** - Процентная доля безработных.

**Urban Population** - Процент населения, проживающего в городских районах.

**Latitude** - Координаты местоположения страны по широте.

**Longitude** - Координаты местоположения страны по долготе.

Далее, построим диаграмму "Ящик с усами" и Гистограмма для визуализации выбросов и определения оптимального метода их удаления.

```python
df['GDP'].plot(kind = 'box', vert = True, title = 'Boxplot для визуализации выбросов GDP', ylabel = 'Значение GDP', xlabel = 'Переменная', fontsize = 10) #ящик с усами для визуализации выбросов
```

Логарифмируем данные и постоим для них гистрограмму 

```python
df['log_GDP'] = np.log(df['GDP'])
df.head()
```

```python
df['log_GDP'].plot(kind='hist', bins=10) 

plt.title("Диаграмма частотностей log_GDP")
plt.xlabel("log_GDP")
plt.ylabel("Частота")

plt.show()
```

Распрпеделние получилось похоже на нормальное.
Выберем способ, с помощью которого мы будем удалять выбросы.

```python
#Метод IQR
q1=df['log_GDP'].quantile(0.25)
q3=df['log_GDP'].quantile(0.75)
iqr = q3-q1

#Находим границы интервала, все значения за пределами которого буду считаться выбросами
low1 = q1 - iqr*1.5 #нижняя границы
up1 = q3 + iqr*1.5 #верхняя граница

#Сделаем фильтрацию и посчитаем те значения, которые являются выбросами
df[(df['log_GDP']<low1) | (df['log_GDP']>up1)].shape[0]
```

```python
#Метод трех сигм
avg = df['log_GDP'].mean()
std = df['log_GDP'].std()
low2 = avg-3*std
up2 = avg+3*std
df[(df['log_GDP']<low2) | (df['log_GDP']>up2)].shape[0]
```

Выбросы мы решили удалить по методу IQR, так как там они были обнаружены. Построим ящик с усами и гистограмму для прологарифмировнной переменной.

```python
df_clean = df[(df['log_GDP']>low1) & (df['log_GDP']<up1)]
df_clean['log_GDP'].plot(kind = 'box', vert = True, title = 'Boxplot для визуализации выбросов log_GDP', ylabel = 'Значение log_GDP', xlabel = 'Переменная', fontsize = 10) 
```

```python
df_clean['log_GDP'].plot(kind='hist', bins=10) 

plt.title("Диаграмма частотностей log_GDP")
plt.xlabel("log_GDP")
plt.ylabel("Частота")

plt.show()
```

Следующим шагом мы построили 4 точечных графика, чтобы посмотреть зависимости различных показателей друг от друга, а также добавили линии тренда, воспользовавшись библиотекой Scikit-learn для машинного обучения.

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
fig, ax = plt.subplots(2, 2, figsize=(14, 13)) # создали полотно для графика из 1 ряда и 2 колонки (1 график)

reg = linear_model.LinearRegression()

ax[0, 0].scatter(df['Minimum wage'], df['Unemployment rate'])
ax[0, 0].set_xlabel('Minimum wage', fontsize=10)
ax[0, 0].set_ylabel('Unemployment rate', fontsize=10)
ax[0, 0].set_title('Точечная диаграмма зависимости безработицы\nот минимальной оплаты труда', fontsize=14)
x = df['Minimum wage'].values.reshape(-1, 1)
y = df['Unemployment rate'].values
reg.fit(x, y)
ax[0, 0].plot(x, reg.predict(x), color='red', linewidth=2)


ax[1, 1].scatter(df['Population'], df['Armed Forces size'])
ax[1, 1].set_xlabel('Population', fontsize=10)
ax[1, 1].set_ylabel('Armed Forces size', fontsize=10)
ax[1, 1].set_title('Точечная диаграмма зависимости численности\nвооруженных сил страны от населения', fontsize=14)
x = df['Population'].values.reshape(-1, 1)
y = df['Armed Forces size'].values
reg.fit(x, y)
ax[1, 1].plot(x, reg.predict(x), color='red', linewidth=2)

ax[0, 1].scatter(df['Physicians per thousand'], df['Out of pocket health expenditure'])
ax[0, 1].set_xlabel('Physicians per thousand', fontsize=10)
ax[0, 1].set_ylabel('Out of pocket health expenditure', fontsize=10)
ax[0, 1].set_title('Точечная диаграмма зависимости затрат\nна здравоохранение от кол-ва врачей на 1 тыс. человек', fontsize=14)
x = df['Physicians per thousand'].values.reshape(-1, 1)
y = df['Out of pocket health expenditure'].values
reg.fit(x, y)
ax[0, 1].plot(x, reg.predict(x), color='red', linewidth=2)

ax[1, 0].scatter(df['Gasoline Price'], df['Co2-Emissions'])
ax[1, 0].set_xlabel('Gasoline Price', fontsize=10)
ax[1, 0].set_ylabel('Co2-Emissions', fontsize=10)
ax[1, 0].set_title('Точечная диаграмма зависимости\nвыбросов СО2 от цены на бензин', fontsize=14)
x = df['Gasoline Price'].values.reshape(-1, 1)
y = df['Co2-Emissions'].values
reg.fit(x, y)
ax[1, 0].plot(x, reg.predict(x), color='red', linewidth=2)

plt.subplots_adjust(wspace=0.4, hspace=0.3)
plt.show()
```

Таким образом, по первому графику можно сказать, что чем ниже минимальная заработная плата, тем выше уровень безработицы в стране. 

На втором графике видим, что чем меньше врачей, тем больше расходы на здравоохранение. 

Третий график: в целом, при разной цене бензина выбросы СО2 не сильно отличаются и могут принимать совершенно разные значения.

Четвертый график: чем больше население, тем выше численность вооруженных сил страны.

```python
# Построение множественной линейной регрессии логарифма ВВП от всех числовых переменных, кроме latitude
import statsmodels.api as sm
X1 = numeric_df[["Land Area(Km2)", 'Armed Forces size', 'Birth Rate','Calling Code','Co2-Emissions', 'Fertility Rate', 'Gross tertiary education enrollment (%)', 'Infant mortality', 'Life expectancy', 'Maternal mortality ratio', 'Minimum wage', 'Physicians per thousand', 'Population', 'Tax revenue (%)', 'Urban_population']] ## независимые переменные
Y1 = numeric_df["log_GDP"] ## целевая переменная
X1 = sm.add_constant(X1) ## добавляем константу

model1 = sm.OLS(Y1, X1).fit() # обучаем модель
```
